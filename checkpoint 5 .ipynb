```python
import math 
class Circle():
    def __init__(self,o_x,o_y,r):
        self.o_x=o_x
        self.o_y=o_y
        self.r=r
       
    def surface(self):
        return round((self.r**2)*(math.pi),3)
    def perimetre(self):
        return round(2*self.r*math.pi,3)
    def isInside(self ,x,y):
        if math.sqrt(((x-self.o_x)**2 + (y-self.o_y)**2))<= self.r:
            return True 
        else:
            return False 
c1=Circle(1,2,4)
print(f'la surface du cercle est {c1.surface()} et de perimetre{c1.perimetre()}')
print('surface=',c1.surface())
print('perimetre=',c1.perimetre())
c1.isInside(0,2)
```

```python
class Bank():
    def __init__(self,balance):
        self.balance=balance
    def deposit(self):
        montant=float(input("Entrer le montant à déposer:"))
        self.balance=self.balance+montant
        print('Montant déposé:',montant)
    def withdraw(self):
        montant=float(input('combien voulez-vous retirer?'))
        if self.balance>=montant:
            self.balance=self.balance-montant
            print('vous avez retirer',montant)
        else:
            print("vous n'avez pas assez de soldes")
    def resume(self):
        print('votre solde est de ',self.balance)
youness=Bank(10000)
youness.deposit()
```
